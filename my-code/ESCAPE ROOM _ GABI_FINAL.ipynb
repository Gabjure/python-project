```python
#Start running here
import time
from termcolor import colored
from playsound import playsound
from PIL import Image 
import os

hc = Image.open("./Hostel Collection.jpeg")
cp = Image.open("./creepy_painting.jpg")
hp2 = Image.open("./hostel_p2.jpg")




#kills kernel
#os._exit(00)

#playsound('./lidcreak.mp3')
#playsound('./headchop.mp3')
#playsound("Scream+Female+2.mp3")
#playsound("close_door_3.mp3")
#def heartbeat():
#    playsound("heartbeat.mp3")
#    playsound("heartbeat.mp3")
#    playsound("heartbeat.mp3")
#    playsound("heartbeat.mp3")

#Image.open("./Hostel Collection.jpeg")

# defines locations
game_room = {
    "name": "Game Room",
    "type": "room",}

bedroom_a = {
    "name": "Bedroom A",
    "type": "room",}

bedroom_b = {
    "name": "Bedroom B",
    "type": "room",}

living_room = {
    "name": "Living Room",
    "type": "room",}

outside = {
  "name": "The Outside"}

death = {
    "name": "Torture Chamber"
}

all_rooms = [game_room, bedroom_a, bedroom_b, living_room, outside, death]

# defines items
couch = {
    "name": "Couch",
    "type": "furniture",
}

piano = {
    "name": "Piano",
    "type": "furniture",
}

queen_bed = {
    "name": "Queen Bed",
    "type": "furniture",
}

double_bed = {
    "name": "Double Bed",
    "type": "furniture",
}

dresser = {
    "name": "Dresser",
    "type": "furniture",
}

dinning_table = {
    "name": "Dinning Table",
    "type": "furniture",
}

big_mirror = {
    "name": "Big Mirror",
    "type": "furniture",
}

small_mirror = {
    "name": "Small Mirror",
    "type": "furniture",
}

creepy_painting = {
    "name": "Creepy Painting",
    "type": "furniture",
}


# defines doors
door_a = {
    "name": "Door A",
    "type": "door",
}

door_b = {
    "name": "Door B",
    "type": "door",
}

door_c = {
    "name": "Door C",
    "type": "door",
}

door_d = {
    "name": "Door D",
    "type": "door",
}

door_e = {
    "name": "Door E",
    "type": "door",
}

all_doors = [door_a, door_b, door_c, door_d, door_e]

# defines keys
key_a = {
    "name": "Key for Door A",
    "type": "key",
    "target": door_a,
}

key_b = {
    "name": "Key for Door B",
    "type": "key",
    "target": door_b,
}

key_c = {
    "name": "Key for Door C",
    "type": "key",
    "target": door_c,
}

key_d = {
    "name": "Key for Door D",
    "type": "key",
    "target": door_d,
}

key_e = {
    "name": "the reflection of death",
    "type": "key",
    "target": door_e
}

hand_gun = {
    "name": "Gun",
    "type": "furniture",
    "target": None
}

hand_gun_bullets = {
    "name": "Bullets",
    "type": "furniture",
    "target": None
}

small_gun = {
    "name": "Small Hand Gun",
    "type": "furniture",
    "target": None
}
# defines which items/rooms are related
object_relations = {
    "Game Room": [couch, piano, creepy_painting, door_a,],
    "Bedroom A": [queen_bed, big_mirror, door_a, door_b, door_c,], 
    "Bedroom B":[double_bed, dresser, small_mirror, door_b,] , 
    "Living Room": [dinning_table, creepy_painting, door_c, door_d, door_e],
    "Torture Chamber": [door_e, hand_gun], #"You should NOT be here, get out of here NOW!"\]
    "The Outside": [door_d],
    "Piano": [key_a],
    "Creepy Painting": [hand_gun_bullets],
    "Queen Bed": [key_b],
    "Double Bed": [key_c],
    "Dresser": [key_d],
    "Big Mirror": [key_e],
    "Small Mirror" : [small_gun],
    "Door A": [game_room, bedroom_a],
    "Door B": [bedroom_a, bedroom_b], 
    "Door C": [bedroom_a, living_room], 
    "Door D": [outside],
    "Door E": [death, living_room]
}

# defines game state. Do not directly change this dict. 
# Instead, when a new game starts, make a copy of this
# dict and use the copy to store gameplay state. This 
# way you can replay the game multiple times.

INIT_GAME_STATE = {
    "current_room": game_room,
    "keys_collected": [],
    "target_room": outside
}

def linebreak():
    """
    Print a line break
    """
    print("\n\n")
    
def getting_shot():
    playsound("Gun+Clip.mp3")
    playsound("Gun+Cock.mp3")
    playsound('./headchop.mp3')    
    
def start_game():
    """
    Start the game
    """
    playsound("heartbeat.mp3")
    print("You wake up on a couch and find yourself in a strange "
          "house with no windows which you have never been to before.\n\n")
    playsound("heartbeat.mp3")
    print("You don't remember why you are here and what had happened before.\n\n")
    playsound("heartbeat.mp3")
    print("You feel some unknown danger is approaching and you must get out of the house...\n\n ")
    playsound("heartbeat.mp3")
    print("You feel uneasy and your heart is racing\n\n")
    playsound("heartbeat.mp3")
    playsound("Scream+Female+2.mp3")
    print("You hear a bang in the distance\n\n")
    playsound('./loud_bang.wav')
    playsound("heartbeat.mp3")
    print("You have to leave as soon as possible\n\n")
    playsound("close_door_3.mp3")
    print(colored("NOW!\n", "red", attrs=["bold"]))
    time.sleep(0.5)
    print(colored("GO!\n", "red", attrs=["bold"]))
    time.sleep(0.5)
    print(colored("GO!\n", "red", attrs=["bold"]))
    time.sleep(0.5)
    print(colored("GO!\n\n", "red", attrs=["bold"]))
    play_room(game_state["current_room"])    
    
def play_room(room):
    """
    Play a room. First check if the room being played is the target room.
    If it is, the game will end with success. Otherwise, let player either 
    explore (list all items in this room) or examine an item found here.
    """
    game_state["current_room"] = room
    if(game_state["current_room"] == game_state["target_room"]):
        playsound('./lidcreak.mp3')
        time.sleep(0.5)
        print(colored("You escape the house!\n", attrs=["bold", "underline"]))
        time.sleep(0.5)
        print(colored("You hear a noise behind you and look back\n", "red", attrs=["bold"]))
        time.sleep(0.5)
        getting_shot()
        print(colored("You've been shot in the head...\n", "red", attrs=["bold"]))
        time.sleep(0.5)
        print(colored("Try escaping faster the next time!\n", "red", attrs=["bold", "blink"]))
        time.sleep(2)
        hc.show()
        print("Do you want to play again?")
        intended_action = input(colored("\n→ yes", attrs=["bold"])
                                + colored("\n→ no\n", attrs=["bold"]) + "\n\n").strip()
        linebreak()
        #UPDATE FOR V. 3
        if intended_action == "yes":
         #   restart_game()
            time.sleep(0.5)
            print(colored("COMING SOON...", "red", attrs=["bold"]))
            time.sleep(0.5)
            hp2.show()
            os._exit(00)
        else:
            print("Your loss")
            os._exit(00)
    elif(game_state["current_room"] == death):
        playsound('./lidcreak.mp3')
        print("You are now in the " + colored(room["name"], attrs=["bold"]) + "!\n")
        time.sleep(1)
        time.sleep(1)
        print(colored("GET OUT AS SOON AS POSSIBLE\n", "red", attrs=["bold"]))
        time.sleep(0.5)
        print("What would you like to do?\n")
        time.sleep(0.5)
        print("Type:")
        time.sleep(0.5)
        intended_action = input(colored("\n→ explore ", attrs=["bold"]) + "\n"
                                + colored("\n→ examine ", attrs=["bold"]) + "\n"
                                + (colored("\n→ GET OUT\n", "red", attrs=["bold"]))).strip()
        if intended_action == "explore":
            time.sleep(0.5)
            explore_room(room)
            time.sleep(0.5)
            play_room(room)
        elif intended_action == "examine":
            time.sleep(0.5)
            examine_item(input("What would you like to " +
                                colored("examine", attrs=["bold"]) + "?\n").strip())
        elif intended_action == "GET OUT":
            cp.show()
            play_room(living_room)
        else:
            print("Not sure what you mean.\n\nType:" + 
                                colored("\n→ explore", attrs=["bold"])+ 
                                colored("\n→ examine\n", attrs=["bold"]))
            play_room(room)
    elif(game_state["current_room"] == game_room):
        time.sleep(0.5)
        print("You are now in the " + colored(room["name"], attrs=["bold"]) + "!\n")
        time.sleep(0.5)
        print("What would you like to do?\n\nType:\n")
        time.sleep(0.5)
        intended_action = input(colored("\n     → explore", attrs=["bold"]) +
                                colored("\n     → examine\n", attrs=["bold"])).strip()
        if intended_action == "explore":
            time.sleep(0.5)
            explore_room(room)
            time.sleep(0.5)
            play_room(room)
        elif intended_action == "examine":
            time.sleep(0.5)
            examine_item(input("\nWhat would you like to " +
                                colored("examine", attrs=["bold"]) + "? ").strip())
        else:
            print("Not sure what you mean.\nType:\n" + 
                                colored("\n→ explore ", attrs=["bold"]) + 
                                colored("\n→ examine\n", attrs=["bold"]))
            play_room(room)
    else:
        playsound('./lidcreak.mp3')
        print("You are now in the " + 
              colored(room["name"], attrs=["bold"]) + 
              "!\n")
        print("What would you like to do?\n\nType:\n")
        time.sleep(0.5)
        intended_action = input(colored("\n→ explore", attrs=["bold"]) + 
                                colored("\n→ examine\n\n", attrs=["bold"])).strip()
        if intended_action == "explore":
            time.sleep(0.5)
            explore_room(room)
            time.sleep(0.5)
            play_room(room)
        elif intended_action == "examine":
            time.sleep(0.5)
            examine_item(input("What would you like to " +
                               colored("examine", attrs=["bold"]) + "?\n").strip())
        else:
            print("\nNot sure what you mean.\nType:\n" + 
                                colored("\n→ explore ", attrs=["bold"]) + 
                                colored("\n→ examine\n", attrs=["bold"]))
            play_room(room)
        linebreak()

def explore_room(room):
    """
    Explore a room. List all items belonging to this room.
    """
    items = [i["name"] for i in object_relations[room["name"]]]
    print("\nYou explore the room.\n")
    time.sleep(0.5)
    print("This is the: " + colored(room["name"], attrs=["bold"]) + "!\n")
    time.sleep(0.5)
    print("You look around and see:\n\n")
    print ("→ a " + "\n\n→ a ".join(items))
    linebreak()
    
def get_next_room_of_door(door, current_room):
    """
    From object_relations, find the two rooms connected to the given door.
    Return the room that is not the current_room.
    """
    connected_rooms = object_relations[door["name"]]
    for room in connected_rooms:
        if(not current_room == room):
            return room
        linebreak()
        
def examine_item(item_name):
    """
    Examine an item which can be a door or furniture.
    First make sure the intended item belongs to the current room.
    Then check if the item is a door. Tell player if key hasn't been 
    collected yet. Otherwise ask player if they want to go to the next
    room. If the item is not a door, then check if it contains keys.
    Collect the key if found and update the game state. At the end,
    play either the current or the next room depending on the game state
    to keep playing.
    """
    current_room = game_state["current_room"]
    next_room = ""
    output = None
    
    for item in object_relations[current_room["name"]]:
        if(item["name"] == item_name):
            output = "\nYou examine the " + colored(item_name, attrs = ["bold"]) + ".\n"
            time.sleep(0.5)
            if(item["type"] == "door"):
                have_key = False
                for key in game_state["keys_collected"]:
                    if(key["target"] == item):
                        have_key = True
                if(have_key):
                    output += "\nYou unlock it with the key you have."
                    next_room = get_next_room_of_door(item, current_room)
                else:
                    output += ("\nIt is locked and you don't have the key.\n\n")
                    time.sleep(0.5)
                    output += colored("Hurry up and find it before your captors return!", "red", attrs = ["bold"])
            else:
                if(item["name"] in object_relations and len(object_relations[item["name"]])>0):
                    if(item["name"] == "Creepy Painting"):
                        item_found = object_relations[item["name"]].pop()
                        game_state["keys_collected"].append(item_found)
                        time.sleep(0.5)
                        output += ("That creeeeepy painting looks like it's watching you...") 
                        time.sleep(0.5)
                        output += ("\nDon't stare at it too long\n")
                        time.sleep(0.5)
                        output += ("You find some " + colored(item_found["name"], attrs=["bold"]) + 
                                   " in a crate under it." )
                    elif(item["name"] == "Big Mirror"):
                        item_found = object_relations[item["name"]].pop()
                        game_state["keys_collected"].append(item_found)
                        time.sleep(0.5)
                        output += ("Wait...\nIs this actually a mirror?\n\nIt's a window!\n" + 
                                    colored("\nHurry HIDE\n", "red", attrs=["bold"]) +
                                    "You see the " + colored(item_found["name"], "red", attrs=["bold"]) + 
                                    ", that can be you next!!" 
                                   + colored("HURRY UP AND GET OUT", "red", attrs=["bold"]))
                    elif(item["name"] == "Small Mirror"):
                        item_found = object_relations[item["name"]].pop()
                        game_state["keys_collected"].append(item_found)
                        time.sleep(0.5)
                        output += ("There's some blood running through your face, " +
                                   "go find something to clean yourself with.")
                        time.sleep(0.5)
                        output += ("You see a " + colored(item_found["name"], attrs=["bold"]) 
                                    + " kicked to a side, take it..\nYou never know what might happen" )
                    else:
                        item_found = object_relations[item["name"]].pop()
                        game_state["keys_collected"].append(item_found)
                        output += "You find " + item_found["name"] + "."                     
                else:
                    if(item["name"] == "Creepy Painting"):
                        output += ("That creeepy painting looks like it's watching you...\n" + 
                        "Don't stare at it too long\n")
                    elif(item["name"] == "Big Mirror"):
                        output += "Did you see that?" + colored("Go hide! Quickly!", attrs = ["bold"])
                    elif(item["name"] == "Small Mirror"):
                        output += "Stop looking at yourself and " + colored("get out", attrs = ["bold"]) + "!"
                    else:
                        time.sleep(0.5)
                        output += ("There isn't anything interesting about it.\n"
                     + colored("Stop messing around or you will get yourself killed", "red", attrs = ["bold"]))
            print(output)
            linebreak()
            break
            
    if(output is None):
        print("You look around but you " +  colored("can't ", "red", attrs=["bold"]) 
              + "see it in the current room.\n")
        time.sleep(0.5)
        print("Are you sure it's here?")
        linebreak()
    
    if(next_room and input("Do you want to go to the next room?\nEnter:\n" + 
                           colored('\n→ yes ', attrs = ["bold"]) + 
                           colored('\n→ no\n\n', attrs = ["bold"])).strip() == 'yes'):
        playsound("close_door_3.mp3")
        linebreak()
        print("Dont't loose the key, You'll need it in case you have to go back\n")
        play_room(next_room)
        linebreak()
    else:
        play_room(current_room)
        linebreak()
```

```python
#RUN TO PLAY THE GAME
game_state = INIT_GAME_STATE.copy()

start_game()
```
